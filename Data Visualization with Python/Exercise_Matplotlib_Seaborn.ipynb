**Q-14** Build a corelation of all predictors with 'Churn' and plot a bar chart (Use dataframe.corr() function)

```python
## Write your code here
df[['Churn']] = df[['Churn']].apply(pd.to_numeric)
df.corr()
```

```python
predictors = []
correlation = []
j = 0

for i in df.corr().columns:
    if i == 'Churn':
        continue
    predictors.append(i)
    correlation.append(df.corr()['Churn'].iloc[j])
    j += 1

predictors
# correlation
```

```python
data = {'predictors': predictors,
        'correlation': correlation}

df_churn = pd.DataFrame(data)
df_churn
```

```python
sns.barplot(data = df_churn, x="predictors", y="correlation")
```
